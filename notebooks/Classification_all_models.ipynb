```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, 1:-1].values
y= df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.2, random_state=0)
```

## **Logistic regression classifier**


```python
## standardizing the x_test and x_train just for the logistic regression
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_trainL = sc.fit_transform(X_train)

X_testL = sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_trainL, y_train)
```

```python
y_test_predictL = classifier.predict(X_testL)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confuse = confusion_matrix(y_test, y_test_predictL)
print(f"consusion matrix is: \n {confuse}")

accurate = accuracy_score(y_test, y_test_predictL)
print(f"accuracy score is: {accurate}")
```

## **K neerest neighbors**

```python
sc1 = StandardScaler()
X_trainKNN = sc1.fit_transform(X_train)
X_testKNN = sc1.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifierKNN = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski', n_jobs=None)
classifierKNN.fit(X_trainKNN, y_train)
```

```python
y_predict_KNN = classifierKNN.predict(X_testKNN)
```

```python
from sklearn.metrics import accuracy_score, pair_confusion_matrix
accuracyKNN = accuracy_score(y_test, y_predict_KNN)
confusionKNN = confusion_matrix(y_test, y_predict_KNN)
print(accuracyKNN)
print(confusionKNN)
```
